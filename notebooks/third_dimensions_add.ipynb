```python
import pandas as pd
```

```python
pop_24 = pd.read_excel('../data/27876_132.xlsx', skiprows=5)
df_commune_responses_combined = pd.read_csv("../data/commune_responses_combined.csv")
df_commune_responses_combined.tail()
```

```python
pop_24.columns = ["Region-ID", "Regionname", "Population", "Density"]
pop_24 = pop_24[["Region-ID", "Density", "Population"]]
pop_24["Region-ID"] = pd.to_numeric(pop_24["Region-ID"], errors="coerce").astype("Int64")
# Clean the data
pop_24["Density"] = pd.to_numeric(pop_24["Density"], errors="coerce")
```

```python
pop_24['Region-ID'] = pop_24['Region-ID'].astype('float')
pop_24.head()
```

```python
df_commune_responses_combined = pd.merge(df_commune_responses_combined, pop_24, left_on = "gemid", right_on="Region-ID", how="left")
```

```python
df_commune_responses_combined.head()
```

```python
df_commune_responses_combined = df_commune_responses_combined.drop_duplicates(subset=['Unnamed: 0'], keep='first')
```

```python
df_commune_responses_combined.tail()
```

```python
df_commune_responses_combined.loc[df_commune_responses_combined['Unnamed: 0'] == 'year', 'gemid'] = 10000
df_commune_responses_combined.loc[df_commune_responses_combined['Unnamed: 0'] == 'quest_glob', 'gemid'] = 20000
```
